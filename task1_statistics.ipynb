```python
import pandas as pd

df = pd.read_csv("region_summary_NSW_cleaned.csv")

print(df.head())
```

# Statistic 1 - Average Resident Population

This statistic calculates the average resident population across NSW regions.

```python
avg_population = df['2024'].mean()

print(avg_population)
```

# Statistic 2 - Maximum Resident Population

This analysis identifies the highest population value in the dataset.

```python
max_population = df['2024'].max()

print(max_population)
```

# Statistic 3 - Minimum Resident Population

This analysis identifies the smallest population value in the dataset.

```python
min_population = df['2024'].min()

print(min_population)
```

The result indicates regions with relatively low population levels.

# Statistic 4 - Population Trend Visualization

This chart visualizes selected population-related values from the dataset.

```python
import matplotlib.pyplot as plt

years = ['2019', '2020', '2021', '2022', '2023', '2024']

values = [
    8046748,
    8110610,
    8097062,
    8166704,
    8341199,
    8479314
]

plt.plot(years, values)

plt.title("NSW Population Trend")

plt.xlabel("Year")

plt.ylabel("Population")

plt.show()
```

The chart shows an overall increase in NSW population over recent years.

# Statistic 5 - Descriptive Statistics

This analysis summarizes the statistical properties of the population dataset.

```python
print(df['2024'].describe())
```

The descriptive statistics provide information such as mean, minimum, maximum and standard deviation.
