```python
import requests
import pandas as pd

url = "https://maps.six.nsw.gov.au/arcgis/rest/services/public/NSW_POI/MapServer/0/query"

params = {
    "where": "1=1",
    "geometry": "150.93,-34.20,151.25,-33.93",
    "geometryType": "esriGeometryEnvelope",
    "spatialRel": "esriSpatialRelIntersects",
    "outFields": "*",
    "returnGeometry": "true",
    "f": "json"
}

response = requests.get(url, params=params)

print(response)

data = response.json()

print(data.keys())


features = data["features"]

print(len(features))
print(features[0])

rows = []


for place in features:

    name = place['attributes']['poiname']

    x = place['geometry']['x']

    y = place['geometry']['y']

    rows.append(["Sutherland", name, x, y])


df = pd.DataFrame(rows, columns=['sa4_name', 'name', 'x', 'y'])

print(df)

df.to_csv('sutherland_poi.csv', index=False)

print("CSV saved")


poi_count = len(df)

summary = pd.DataFrame({
    "sa4_name": ["Sutherland"],
    "poi_count": [poi_count]
})

print(summary)

summary.to_csv("Sutherland_poi_summary.csv", index=False)

print("Summary CSV saved")
```
